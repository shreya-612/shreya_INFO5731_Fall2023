```python
# You code here (Please add comments in the code):
# Importing necessary libraries
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
import json

# Defining the URL of the IMDb movie reviews page
movies_url = "https://www.imdb.com/title/tt6751668/reviews?ref_=tt_urv"

# Opening the URL and parsing its contents using BeautifulSoup
pages = urllib.request.urlopen(movies_url)
soup = BeautifulSoup(pages)

# Initializing lists to store data from the HTML page
UserName = [] # Usernames of reviewers
star = [] # Star ratings given by reviewers
review_title = []  # Titles of reviews
review_text = []  # Review text
review_posted_time = [] # Timestamps when reviews were posted

# Extracting data from the HTML page using BeautifulSoup

# Extracting usernames of reviewers
for user in soup.find_all("span",{'class':'display-name-link'}, 'a'):
  user = UserName.append(user.text.strip())

# Extracting star ratings given by reviewers
for Rating in soup.find_all("span",{'class':'rating-other-user-rating'},'span'):
  Rating = star.append(Rating.text.strip())

# Extracting titles of reviews
for title in soup.find_all("a",{'class':'title'}):
  title = review_title.append(title.text.strip())

# Extract review text
for reviewtext in soup.find_all("div", {'class':'text show-more__control'}):
  reviewtext = review_text.append(reviewtext.text.strip())

# Extracting timestamps when reviews were posted
for postedtime in soup.find_all("span",{'class':'review-date'}):
  postedtime = review_posted_time.append(postedtime.text.strip())

# Creating a dictionary to store the extracted data
rev = {'User Name':UserName, 'Ratings':star,'Review title':review_title, 'Review':review_text, 'Review Posted Time':review_posted_time}

# Creating a pandas DataFrame from the dictionary
reviews_1 = pd.DataFrame.from_dict(rev, orient='index')

# Transposing the DataFrame to have rows as individual reviews and columns as attributes
reviews_1.transpose()
```
