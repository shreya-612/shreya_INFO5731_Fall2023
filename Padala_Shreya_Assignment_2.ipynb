```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import re

# Creating an empty list to store the collected book reviews.
harrypotter_book_reviews=[]

# Looping through multiple pages to collect reviews for Harry Potter and the Prisoner of Azkaban book.
for page_number in range(1,1000):
  page_reviews=[]  # Creating an empty list to store reviews on this page.

  # Defining the URL for the Amazon product reviews page, updating the page number.
  url="https://www.amazon.com/product-reviews/1338815288/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&filterByStar=all_stars&reviewerType=all_reviews&pageNumber=1#reviews-filter-bar"+str(page_number)

  # Sending an HTTP GET request to the URL.
  response = requests.get(url)

  # Parsing the HTML content of the page using BeautifulSoup.
  soup = bs(response.content,"html.parser")

  # Finding and collecting review elements on the page.
  reviews = soup.find_all("div",attrs={"class","a-row a-spacing-small review-data"})

  # Iterating through the found reviews on the page and removing newline characters.
  for page_number in range(len(reviews)):
    page_reviews.append((reviews[page_number].text).replace('\n',''))
    harrypotter_book_reviews=harrypotter_book_reviews+page_reviews

# Creating a Pandas DataFrame to store the collected reviews with a meaningful column name and displaying the resulting DataFrame.
reviews= pd.DataFrame(harrypotter_book_reviews,columns=['reviews'])
reviews
```

```python
reviews.to_csv('bookreviews.csv', index=False)
```

```python
# Write your code here
import pandas as pd
# Reading data from a CSV file named "bookreviews.csv" and storing it in a DataFrame.
reviews_df = pd.read_csv("bookreviews.csv")
# Displaying the contents of the DataFrame, showing the data from the CSV file.
reviews_df
```

```python
# Adds a new column to the DataFrame to store the number of characters in each review.
reviews_df['No. of Characters'] = reviews_df['reviews'].str.len()
# Displays the DataFrame with the new column.
reviews_df
```

```python
# Adding a new column to the DataFrame to store the number of numeric words in each review.
reviews_df['No. of numerics'] = reviews_df['reviews'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
# Displaying the DataFrame with the new column.
reviews_df
```

```python
# Adding a new column to the DataFrame to store the review text with digits removed.
reviews_df['removal of digits'] = reviews_df['reviews'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
# Displaying the DataFrame with the new column.
reviews_df
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Import the necessary NLTK library.
from nltk.corpus import stopwords
# Defining the list of English stopwords.
stop = stopwords.words('english')
# Adding a new column to the DataFrame to store the review text with stopwords removed.
reviews_df['removal of stopwords'] = reviews_df['reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
# Displaying the DataFrame with the new 'removal of stopwords' column.
reviews_df
```

```python
# Adding a new column to the DataFrame to store the review text in lowercase.
reviews_df['text_lowercase'] = reviews_df['reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))
# Displaying the DataFrame with the new 'text_lowercase' column.
reviews_df
```

```python
# Importing the PorterStemmer from NLTK.
from nltk.stem import PorterStemmer
# Initializing the PorterStemmer.
st = PorterStemmer()
# Adding a new column to the DataFrame to store the stemmed review text.
reviews_df['stemmed_reviews'] = reviews_df['text_lowercase'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
# Displaying the DataFrame with the new 'stemmed_reviews' column.
reviews_df
```

```python
# Adding a new column to the DataFrame to store the review text with punctuation removed.
reviews_df['punctuation_free_text'] = reviews_df['text_lowercase'].str.replace('[^\w\s]', '')
# Displaying the DataFrame with the new 'punctuation_free_text' column.
reviews_df
```

```python
import spacy
import nltk
from nltk import pos_tag, ne_chunk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tree import Tree
from collections import Counter

# Load the spaCy English model for dependency parsing and named entity recognition.
nlp = spacy.load("en_core_web_sm")

# Function to print constituency parsing tree using NLTK.
def print_constituency_tree(text):
  """Prints the constituency parsing tree of a text using NLTK.
  Args:
    text: The text to parse.
  Returns:
    None.
  """
  sentences = nltk.sent_tokenize(text)
  for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    tagged = pos_tag(words)
    chunked = ne_chunk(tagged)
    for subtree in chunked:
      if type(subtree) == Tree:
        print(subtree.label(), " ".join(word for word, pos in subtree.leaves()))
      else:
        print(subtree[0], subtree[1])

# Function to print dependency parsing tree using spaCy.
def print_dependency_tree(text):
  """Prints the dependency parsing tree of a text using spaCy.
  Args:
    text: The text to parse.
  Returns:
    None.
  """

  doc = nlp(text)
  for token in doc:
    print(token.text, token.dep_, token.head.text, [child for child in token.children])

# Function to extract named entities and count their occurrences.
def extract_named_entities(text):
  """Extracts named entities from a text and counts their occurrences.
  Args:
    text: The text to extract named entities from.
  Returns:
    A Counter object containing the named entities and their counts.
  """

  doc = nlp(text)
  entity_counter = Counter()
  for ent in doc.ents:
    entity_counter[ent.label_] += 1
  return entity_counter

# Load the cleaned text from the CSV file.
with open("cleaned_bookreviews.csv", "r") as file:
  text = file.read()

# (1) Parts of Speech (POS) Tagging
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

# Count the number of nouns, verbs, adjectives, and adverbs in the text.
noun_count = len([word for word, pos in pos_tags if pos.startswith('N')])
verb_count = len([word for word, pos in pos_tags if pos.startswith('V')])
adj_count = len([word for word, pos in pos_tags if pos.startswith('J')])
adv_count = len([word for word, pos in pos_tags if pos.startswith('R')])

print(f"Total Nouns: {noun_count}")
print(f"Total Verbs: {verb_count}")
print(f"Total Adjectives: {adj_count}")
print(f"Total Adverbs: {adv_count}")

# (2) Constituency Parsing and Dependency Parsing
print("Constituency Parsing Trees:")
print_constituency_tree(text)
print("\nDependency Parsing Tree:")
print_dependency_tree(text)

# (3) Named Entity Recognition
entity_counter = extract_named_entities(text)
print("Named Entities:")
for entity, count in entity_counter.items():
  print(f"{entity}: {count}")
```

**Write your explanations of the constituency parsing tree and dependency parsing tree here (Question 3-2):**


Constituency parsing tree: A tree that depicts the hierarchical structure of a sentence is called a constituency parsing tree, sometimes referred to as a phrase structure tree. Its foundation is the notion that sentences can be disassembled into ever-tinier components, like words, phrases, and clauses. The top-level ingredient, usually the sentence itself, is the root of the tree. The lower-level elements are represented by the other nodes in the tree, which are connected to the root node by edges.

Dependency parsing tree: The syntactic connections between the words in a phrase are represented by a dependency parsing tree, also called a dependency graph. It is predicated on the notion that all words, with the exception of the sentence's root word, are dependent on one another in sentences. Usually the verb, the root word is the most autonomous word in the sentence.
